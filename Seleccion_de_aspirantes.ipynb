```python
# función para calcular valor de puntaje para admisión,
# formula según el departamento
def calcular_puntaje_admision(linea, opcion):
  departmento = linea[opcion]

  if departmento == 'Physics':
    puntajes = [linea['EPH'], linea['EMA'], linea['ESP']]
  elif departmento == 'Mathematics':
    puntajes = [linea['EMA'], linea['ECS'], linea['ESP']]
  elif departmento == 'Engineering':
    puntajes = [linea['EMA'], linea['ECS'], linea['EPH'], linea['ESP']]
  elif departmento == 'Chemistry':
    puntajes = [linea['ECH'], linea['EPH'], linea['ESP']]
  elif departmento == 'Biotech':
    puntajes = [linea['EMA'], linea['ECS'], linea['ECH'], linea['ESP']]
  else:
    return None  # Para caso cuando departamento no esta en la lista

  return round(sum(puntajes) / len(puntajes), 2)
```

```python
# Determinar la mejor opción para cada aspirante
def mejor_opcion(row):
  scores = {'OP1': row['OP1_score'], 'OP2': row['OP2_score'], 'OP3': row['OP3_score']}
  mejor_op = max(scores, key=scores.get)
  return row[mejor_op]

df['mejor_opcion'] = df.apply(mejor_opcion, axis=1)

# Generar el mensaje para cada aspirante
def generar_mensaje(row):
  mensaje = f"Estimado/a {row['nombre_completo']}: Según su puntaje, su mejor opción es {row['mejor_opcion']}. Le sugerimos que consulte la capacidad disponible para esta opción."
  return mensaje

df['mensaje_aspirante'] = df.apply(generar_mensaje, axis=1)

# Mostrar los mensajes para los primeros aspirantes
for index, row in df.head().iterrows():
  print(row['mensaje_aspirante'])
```
